# CLEANING OF DATA

```python
backup=car.copy() #used for larger dataset
```

```python
car['year']
```

```python
car['year'].str.isnumeric() #basically it tells whether the columns has numeric value or not , if it has numeric value then o/p will be True
```

```python
car=car[car['year'].str.isnumeric()] #only that will show which has numeric values
```

```python
car['year']=pd.to_numeric(car['year'],errors='coerce').fillna(0).astype(int)
```

```python
car['year']
```

```python
car.info()
```

```python
car=car[car['Price']!="Ask For Price"].copy() # it'll remove the column which has float value
```

```python
car['Price']=car['Price'].str.replace(',','') #it will remove comma 
```

```python
car['Price']=car['Price'].astype(int)
```

```python
car['Price']
```

```python
car['kms_driven']
```

```python
car['kms_driven'].str.split(' ')
```

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car['kms_driven']
```

```python
car['kms_driven']=pd.to_numeric(car['kms_driven'],errors='coerce')
```

```python
car=car.dropna(subset=['kms_driven'])
```

```python
car['kms_driven']
```

```python
car.info()
```

```python
car[car['fuel_type'].isna()]
```

```python
car=car[~car['fuel_type'].isna()]
```
